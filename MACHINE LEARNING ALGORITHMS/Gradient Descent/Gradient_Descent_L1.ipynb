# GRADIENT DESCENT

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4 , n_features=1 , n_informative=1 , n_targets=1 , noise=80 , random_state=13) 
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Lets apply OLS 
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# Lets apply gradient descent assuming slope is constant m = 78.35
# and lets's assume the starting value for intercept b =0

y_pred = ((78.35 * x) + 0 ).reshape(4)
# calculating all y values for a line which is incorrect right now
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
# lets start calculation
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m * x.ravel() - b )
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1

# n*slope
step_size = loss_slope * lr
step_size
```

```python
# calculate the new intercept 
# b old = b new - n*slope
b1 = b - step_size
b1
```

```python
y_pred1 = ((78.35 * x) + b1).reshape(4)
```

```python
# Let's plot a graph to show new b line
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='black',label='b1 = {}'.format(b1))
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum( y - m * x.ravel() - b1)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b2 = b1 - step_size
b2
```

```python
y_pred2 = ((78.35 * x) + b2).reshape(4)
```

```python
# Let's plot a graph to show new b line
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='black',label='b1 = {}'.format(b1))
plt.plot(x,y_pred2,color='#A3E4D7',label='b2 = {}'.format(b2))
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m * x.ravel() - b2)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b3 = b2 - step_size
b3
```

```python
y_pred3 = ((78.35 * x) + b3).reshape(4)
```

```python
# Let's plot a graph to show new b line
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='black',label='b1 = {}'.format(b1))
plt.plot(x,y_pred2,color='#A3E4D7',label='b2 = {}'.format(b2))
plt.plot(x,y_pred3,color='gold',label='b3 = {}'.format(b3))
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
# The beauty of Gradient decent is is that if we take any value of b (let b = 100) the it will reach the minimum value of 
# loss function by taking longer steps
```

```python
# combined code for Gradient Descent
b = -100
m = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum( y - m * x.ravel() - b)
    b = b - loss_slope * lr
    
    y_pred = m * x + b
    
    plt.plot(x,y_pred)
    
plt.scatter(x,y)

# Here our answer depends on lr and epochs as on taking small lr it is possible that we cannot reach correct answer
# now if we increase epochs then only we can reach correct answer
# concluding : if we keep learning rate low we can reach correct answer but for high learning rate there would be some mistakes
```
