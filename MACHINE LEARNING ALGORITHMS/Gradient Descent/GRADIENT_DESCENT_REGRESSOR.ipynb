```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr.fit(x,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
m = 35.65
```

```python
class GDRegressor:
    def __init__(self,learning_rate,epochs):
        self.m = 35.65
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,x,y):
        # caclulate the b using GD
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - self.m * x.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
            print(loss_slope,self.b)
            
        print(self.b)
            
```

```python
gd = GDRegressor(0.01,10)
```

```python
gd.fit(x,y)
# here we can see that we choose a learning rate of 0.01 which make a zig zag pattern in the graph of the regression line
# so to avoid this we choose a learning rate of 0.001( or say reduce ) and incraese the epochs (say 100)
```

```python
GD = GDRegressor(0.001,100)
```

```python
GD.fit(x,y)
```
